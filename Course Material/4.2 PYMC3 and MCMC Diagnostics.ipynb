## Definitions

+ ** Probabilistic Programming **  is the field which unifies general purpose modelling with probabilistic modeling[1]. It helps create systems which help make decisions when uncertainty is present. Probabilistic reasoning combines knowledge of a situation with the laws of probability. Till a few years back, probabilistic modeling had limited use because of their complex calculations. The invent of probabilistic programming has made this approach more widespread by making it easier to build models and get inferences.
+ ** Probabilistic Programming Languages (PPL) ** are high level programming languages which have been designed to define probabilistic models followed by automatically solving those models[2]. They are very expressive and adjustable. Some examples of these are - Stan, PyMC3, R2, Edward

### Benefits of Probabilistic Programming

Probabilistic Programming languages do all the heavy lifting for us through their inference algorithms, which keep on readjusting the associated probabilities every time new data comes in. This can vastly reduce the time and effort associated with implementing new models and understanding data.

## Markov Chain Monte Carlo (MCMC)
A sequence $X_1$, $X_2$, ... of random elements of some set is a **Markov chain** when the conditional
distribution of $X_{n+1}$ given $X_1$, ... , $X_n$ depends on $X_n$ only.

The joint distribution of a Markov chain is determined by:<br>
• The marginal distribution of $X_1$, called the initial distribution<br>
• The conditional distribution of $X_{n+1}$ given $X_n$, called the transition probability distribution[7]<br>

In Bayesian statistics, our aim is to calculate the posterior distribution, which is very complex due to the high-dimensional integral (marginal likelihood). We can approximate the posterior $p(θ\mid X)$ if we can somehow draw many samples that come from the posterior distribution.

In **MCMC**, we draw samples from a simple distribution such that every sample relies only on the state of preceding draw (property of Markov chain). Furthermore, all samples are not used as we setup acceptance standards for each draw primarily based on comparing successive states with respect to a target distribution that make sure that the stationary distribution is the posterior distribution of interest. Another good thing is that the target distribution effectively needs to be only proportional to the posterior distribution, so that we don’t need to calculate the complex marginal likelihood.
$$
posterior \propto likelihood \times prior
$$

#### Calculation of Acceptance ratio
For a Stable Markov chain we have,
\begin{align}
p(x)p(x \to x') =& p(x')*p(x \to x') \\
\\
\text{When moving from stage  }.. x_t \to x_t' \\
\\
\frac{p(x_t \to x_t')}{p(x_t' \to x_t)} &= \frac{p(x_t')}{p(x_t)} \tag{1} \\\
\end{align}

Probability of transition from state $x$ to $x'$
\begin{align}
P(x \to x') &= g(x \to x')*A(x \to x') \tag{2}\\
&= \text{proposal} * \text{acceptance}\\
\\
\text{Using equation (1) and (2)} \\
\\
\frac{g(x_t \to x_t')*A(x_t \to x_t')}{g(x_t' \to x_t)*A(x_t' \to x_t)} &= \frac{p(x_t')}{p(x_t)}
\end{align}

$$
\text{Acceptance ratio } = r = \frac{A(x_t \to x_t')}{A(x_t' \to x_t)} = \frac{p(x_t')*g(x_t \to x_t')}{p(x_t)*g(x_t' \to x_t)}
$$

We move from $x_t$ to state $x_{t+1}$, conditionally based on r as:
+ If $r \geq 1$, then $x_{t+1}=x_t'$
+ If $r \leq 1$, then $x_{t+1}=x_t'$ with Probability r and $x_{t+1}=x_t$ with probability (1-r)

**The basic steps involved in running a MCMC algorithm are as bellow:** 
+ First we select initial values of the input parameters to the model
+ Then we assign new values to the input parameters according to current state
+ If the new values from previous step agree with the observations(i.e. calculate the acceptance ratio), accept them as the new values. If they do not agree with the observation, then we reject them and return to the last state
+ Steps 2 and 3 are repeated till end conditions is reached.
We will now use Metropolis-Hastings algorithm for MCMC and implement an example. We will then run diagnostic tests on it.

We will see an example of how MCMC works in the code below

```python
import os
import sys
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats

%matplotlib inline
```

```python
def target(lik, prior, n, h, theta):
    if theta < 0 or theta > 1:
        return 0
    else:
        return lik(n, theta).pmf(h)*prior.pdf(theta) # calculate prior * likelihood

n = 100
h = 61
a = 10
b = 10
lik = stats.binom
prior = stats.beta(a, b)
sigma = 0.3

naccept = 0
theta = 0.1
niters = 100
samples = np.zeros(niters+1)
samples[0] = theta
sample_table=pd.DataFrame(columns=['Current Value','Candidate','Acceptance Ratio','u','Accept'])
for i in range(niters):
    theta_p = theta + stats.norm(0, sigma).rvs()
    rho = min(1, target(lik, prior, n, h, theta_p)/target(lik, prior, n, h, theta )) # calculating acceptance ratio
    u = np.random.uniform() # generating a uniform random number betweeen 0 and 1
    if u < rho:    # accepting the sample if accceptance ratio is more than the random number generated
        sample_table = sample_table.append({'Current Value': theta, 'Candidate': theta_p, 'Acceptance Ratio': rho,'u':u,'Accept':'YES'},ignore_index = True)
        naccept += 1
        theta = theta_p
    if u >= rho:    #  # rejecting the sample if accceptance ratio is less than the random number generated
        sample_table = sample_table.append({'Current Value': theta, 'Candidate': theta_p, 'Acceptance Ratio': rho,'u':u,'Accept':'NO'},ignore_index = True)
    samples[i+1] = theta
nmcmc = len(samples)//2
print("Efficiency = ", naccept/niters)
```

Efficiency of 0.23 means that out of 100 only 23 proposed candidates have been accepted.

```python
plt.plot(samples)
```

```python
sample_table[:10]
```

The above table shows each step (iteration) in the MCMC chain run in the code earlier.
+ **Current value** is the value using which it samples new values(candidates). It was initialized to 0.1 in our code and then changes as the chain moves forward. It is the current state in Markov Chain.
+ **Candidate** is the random sample generated by the algorithm. It is the proposed next state in Markov Chain.
+ **Acceptance Ratio** is the calculated by using the formula shown in the earlier cells
+ **"u"** is the random number generated from a uniform distribution between 0 and 1
+ **Accept** Column tells us whether the proposed candidate was accepted or not. It is inferred by comparing the Acceptance ratio and the random number generated(u). It is based on rules mentioned earlier in the MCMC explanation

First column of table tells us that transition from initial state (0.10) to proposed state/sample ($-1.14 * 10^{-01}$) was not accepted as the acceptance ratio is less than the random number generated. So in the second step also, the current value (state) will remain same as 0.10.
We see that the proposed candidate(state) has been accepted in step 4 as the acceptance ratio is 1. In this case the proposed value acts as the current value (or state) in the next step, as can be seen from the table

## MCMC implementation and Diagnostics

We will now implement MCMC and perform diagnosis on the obtained results[12].
Suppose we have a two-dimensional set of N points $\{x_i, y_i\}$. The model **M** predicts the values of y as being
$$
y = f(x) + \epsilon \\
f(x) = b_0 + b_1x
$$

The error($\epsilon$) is zero-mean Gaussian random variable with standard deviation $\sigma$, i.e. $\epsilon ∼ N(0, σ)$.

```python
## ============================ Importing necessary packages==================================
import pymc3 as pm
import seaborn as sns
palette = 'muted'
sns.set_palette(palette); sns.set_color_codes(palette)
np.set_printoptions(precision=2)
pd.set_option('display.precision', 2)
```

```python
###====================== Generated simulated data for the model==============================
N = 100
b0_r = 0
b1_r = 1
eps_r = np.random.normal(0, 1, size=N)

x = np.random.normal(3, 1, N)
y_real = b0_r + b1_r * x 
y = y_real + eps_r
```

```python
####========================= Plotting the simulated data=====================================
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.plot(x, y,'b.' )
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$y$', fontsize=16)
plt.plot(x, y_real,'k')
plt.title("X Vs Y with and without noise")
plt.subplot(1,2,2)
sns.kdeplot(y)
plt.xlabel('$y$', fontsize=16)
plt.title("Density function of Y")
plt.tight_layout()
plt.savefig('x_y_yreal.png', dpi=300)
```

+ From the above graphs, we can see that the Y-intercept of the regression line will almost every time lie between 0 and 2. Therefore, we define the variable $b_0$ with priors $\mu = 0$, and $\sigma = 2$
+ The variable $b_1$ in the regression equation is the slope of the line. We can see from the data points the slope can be between $\tan 0$ and $\tan 80$. This corresponds to range (0,9). We will take uniform prior for this variable with the specified upper and lower limits
+ The noise prior will be defined as a function of Half Normal function

```python
###=================== Define the model and draw the trace and density charts==========================
with pm.Model() as model:
    b0 = pm.Normal('b0', mu=0, sd=2)
    b1 = pm.Uniform('b1', lower=0, upper=6)
    #epsilon = pm.HalfCauchy('epsilon', 5)
    epsilon = pm.HalfNormal('epsilon', sd=1)
    
    fx = pm.Deterministic('fx', b0 + b1 * x)
    y_pred = pm.Normal('y_pred', mu=fx, sd=epsilon, observed=y)
    
    start = pm.find_MAP() 
    step = pm.Metropolis() 
    trace = pm.sample(10000, step, start)
```

```python
####=================================== Density and Trace charts==========================================
pm.traceplot(trace)
```

We see from above charts that the trace plot shows random behavior which is a good sign. We have 2 chains for each variable ($b_0,b_1$ and epsilon) and these chains are showing converging properties. **Autocorrelations** between the samples returned by the MCMC algorithm is another way to look at the convergence of chains. he lag-k autocorrelation is the correlation between every sample and the sample k steps before. This should become smaller as the chain progresses. If it remains high throughout the run, that means our samples ae not random and there is a high correlation between them. If we face autocorrelation, we can **thin the MCMC chain** by discarding _n_ samples for every sample that we keep. 

In our example we see that the autocorrelation plot decreases as we move along and this means we have random samples which is a good thing.

#### GEWEKE DIAGNOSTIC

It is a time-series approach that compares the mean and variance of segments from the beginning and end of a single chain.

$$
Geweke Score = z = \frac{E[\, x_s]\,- E[\,x_e]\,}{\sqrt{V[\, x_s] + V[\, x_e]}}
$$

E is the mean, V is variance, $x_s$ is the section at the start of the series and $x_e$ is the section at the end of the series.
**If the chain has converged properly, most of the geweke scores should fall within 2 standard deviations of zero.**

```python
geweke_score = pm.geweke(trace,first = 0.2,last = 0.2,intervals = 20)
```

```python
print(trace.varnames)
b0=geweke_score[0]
```

```python
score=pm.geweke(trace, first=0.1, last=0.5, intervals=20)
plt.scatter(b0['b0'][:,0],b0['b0'][:,1], marker = 'o', s=100)
plt.axhline(-1.98, c='r')
plt.axhline(1.98, c='r')
plt.ylim(-2.5,2.5)
plt.xlim(0-10,trace['b0'].shape[0]/2+10)
plt.title('Geweke Plot Comparing first 10% and Slices of the Last 50% of Chain\nDifference in b0 score')
plt.show()
```

```python
score=pm.geweke(trace, first=0.1, last=0.5, intervals=20)
plt.scatter(b0['b1'][:,0],b0['b1'][:,1], marker = 'o', s=100)
plt.axhline(-1.98, c='r')
plt.axhline(1.98, c='r')
plt.ylim(-2.5,2.5)
plt.xlim(0-10,trace['b1'].shape[0]/2+10)
plt.title('Geweke Plot Comparing first 10% and Slices of the Last 50% of Chain\nDifference in b1 score')
plt.show()
```

We see from above that, all the geweke scores lie in between 2 standard deviations from the zero. So our model is good.

#### Effective Sample Size (ESS)
This quantity estimates the number of independent samples obtained from a run for a particular parameter. If there are some correlation between successive samples in x, then we might expect that our sample has not revealed as much
information of the posterior distribution of our parameter as we could have gotten if the samples were independent.

In practice, we discard some predetermined burn-in value (e.g. 10%, 25% or 50% of the total chain length). Then, it is checked if the ESS value is greater than some predetermined threshold value for the remaining samples. If yes, then we accept the simulation length as sufficient.

```python
pm.diagnostics.effective_n(trace, varnames=varnames, include_transformed=False)
```

#### GELMAN-RUBIN Diagnostic

It tests for non-convergence by comparing the variance between multiple chains to the variance within each chain. If convergence has been achieved, the between-chain and within-chain variances should be identical. To be most effective in detecting evidence for nonconvergence, each chain should be initialized to dispersed starting values relative to the desired distribution. It can also be seen very easily by drawing traceplots of multiple chains and looking for their convergence.
Formula:
$$
\text{Gelman-Rubin score} = R=\frac{V}{W}
$$
where W is the within-chain variance and V is the posterior variance estimate for the pooled traces.

```python
###=================== Define the model and draw the trace and density charts==========================
with pm.Model() as model:
    b0 = pm.Normal('b0', mu=0, sd=2)
    b1 = pm.Uniform('b1', lower=0, upper=6)
    #epsilon = pm.HalfCauchy('epsilon', 5)
    epsilon = pm.HalfNormal('epsilon', sd=1)
    
    fx = pm.Deterministic('fx', b0 + b1 * x)
    y_pred = pm.Normal('y_pred', mu=fx, sd=epsilon, observed=y)
    
    #start = pm.find_MAP() 
    step = pm.Metropolis() 
    multi_trace = pm.sample(10000, step,njobs=10)
```

```python
pm.traceplot(multi_trace)
```

```python
pm.diagnostics.gelman_rubin(multi_trace, varnames=varnames, include_transformed=False)
```

+ We can observe the traceplots above that all the different chains seem to converge together 
+ The geman-rubin diagnostics for all variables is also close to 1, which is a good sign about the convergence of the chains.
