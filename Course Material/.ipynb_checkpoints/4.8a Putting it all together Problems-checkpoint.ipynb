**(b)**  Fit the model with Mixed effect model using maximum likelihood as we did in assignment.<br>

$$Y_{i1k} = b_i + b_{i1} + {b^*}_{ik} + e_{i1k}$$ 
<br>
$$Y_{i2k} = \beta_0+ b_i + b_{i2} + {b^*}_{ik} + e_{i2k}$$  

It can be composed of a fixed effect, and the error term was to represent the deviations from our predictions due to “random” factors that we cannot control experimentally. And ${b^*}_{ik}$ and the error are the “probabilistic” or “stochastic” part of the model.
In mixed effect model, we change the random aspect of our model, essentially leaving the systematic part unchanged. And everything in the “systematic” part of your model works just like with linear models.<br>
And the model linked data follows the distribution:<br>

$$\begin{pmatrix} \bar Y_{i_1}\\\bar Y_{i_1}\end{pmatrix} = \begin{pmatrix} \mu_b \\ \mu_b+ \beta_{0} \end{pmatrix}$$
<br>
$$\begin{pmatrix} {S_1}^2 & S_{12} \\ S_{12} & {S_2}^2 \end{pmatrix} = \begin{pmatrix} \sigma_b^2+\psi^2+\sigma_{b^*}^2+\sigma_{e_1}^2 & \sigma_b^2+\sigma_{b^*}^2 \\ \sigma_b^2+\sigma_{b^*}^2 & \sigma_b^2+\psi^2+\sigma_{b^*}^2+\sigma_{e_2}^2  \end{pmatrix}$$

```python
data=pd.DataFrame(columns=['IC','RV'])

data['IC'] = A
data['RV'] = B

data=data.reset_index(drop=True)

md=smf.mixedlm("IC ~ RV", data, groups=data["RV"])
mdf=md.fit()

m_A=np.mean(A)
m_B=np.mean(B)
Var_A=np.var(A)
Var_B=np.var(B)
Covarm_AB=np.cov(A,B)
Covar_AB=Covarm_AB[0][1]
Var_A=Covarm_AB[0][0]
Var_B=Covarm_AB[1][1]

print('The mean of IC is ',m_A)
print('The mean of RV is ',m_B)
print('The Variance of IC is ',Var_A)
print('The Variance of RV is ',Var_B)
print('The Covariance of IC and RV is ',Covar_AB)
print(' ')
print(mdf.summary())
```

The covariance between these two method is 1.26, which indicates there can be a certain agreement or accuracy between the existing method. The mixed-effect model gives the fitting results as the the table shown above. We can see the Maximum Likelihood optimization converged, and successfully to fit the mixed linear model.

**(d)** Evaluate agreement between the methods. Do the methods agree sufficiently well to be used interchangeably? If not, is there any recalibration that may bring them closer.
Agreemnet for lined paired model:
$$CCC=\frac{\sigma_b^2+\sigma_{b^*}^2}{\beta_0^2+2(\sigma_b^2+\psi^2+\sigma_{b^*}^2)+\sigma_{e_1}^2+\sigma_{e_2}^2 }$$
<br>
$$TDI_{(p)}=\{ (2\psi^2+\sigma_{e_1}^2+\sigma_{e_2}^2)\chi_{1,p}^2(\beta_0^2/\{2\psi^2+\sigma_{e_1}^2+\sigma_{e_2}^2\})\}^{1/2}$$

```python
CCC=(2*(Covar_AB))/((m_A-m_B)**2+Var_A+Var_B)
D=np.abs(A-B)
TDI=np.percentile(D, (100*0.95))
print('CCC before calibration', CCC)
print('TDI before calibration', TDI)
```

```python
diff_m = m_A-m_B
A_recalib = A-diff_m

data_cali=pd.DataFrame(columns=['IC','RV'])

data_cali['IC'] = A_recalib
data_cali['RV'] = B

data_cali=data_cali.reset_index(drop=True)

md=smf.mixedlm("IC ~ RV", data_cali, groups=data_cali["RV"])
mdf=md.fit()

m_A=np.mean(A_recalib)
m_B=np.mean(B)
Var_A=np.var(A_recalib)
Var_B=np.var(B)
Covarm_AB=np.cov(A_recalib,B)
Covar_AB=Covarm_AB[0][1]
Var_A=Covarm_AB[0][0]
Var_B=Covarm_AB[1][1]

print('After Calibration:')
print('The mean of IC is ',m_A)
print('The mean of RV is ',m_B)
print('The Variance of IC is ',Var_A)
print('The Variance of RV is ',Var_B)
print('The Covariance of IC and RV is ',Covar_AB)
print(' ')
print(mdf.summary())
```

```python
CCC=(2*(Covar_AB))/((m_A-m_B)**2+Var_A+Var_B)
D=np.abs(A-B)
TDI=np.percentile(D, (100*0.95))
print('After Calibration:')
print('CCC before calibration', CCC)
print('TDI before calibration', TDI)
```

After calibration, these method became zero-mean, and the CCC inmproved (CCC from 0.66 to 0.73), which means the agreement indeed imporved a little bit. But the TDI score still remains the same (2.11).

**(b)** Bootstrapping, construct 90% confidence intervals for median and standard deviation. In this problem, I chose the bootstrap resamples size as 1000.

```python
n = len(Speed)
B = 1000

mean = np.empty(B)
std =np.empty(B)
med =np.empty(B)

for i in range(B):
    random_sample=np.random.choice(Speed,n)
    sample_mean=np.mean(random_sample)
    sample_std=np.std(random_sample)
    sample_med=np.median(random_sample)
    
    mean[i]=sample_mean
    std[i]=sample_std
    med[i]=sample_med

print("Bootstrap estimated mean  : ", np.mean(mean))
print("Bootstrap estimated std   : ", np.mean(std))
print("Bootstrap estimated median: ", np.mean(med))
print(" ")

Confidence=90
C_low=(100-Confidence)/2
C_high=100-C_low
C_interval_mean=np.percentile(mean,[C_low,C_high])
C_interval_std=np.percentile(std,[C_low,C_high])
C_interval_med=np.percentile(med,[C_low,C_high])

print('90% CI of mean:',C_interval_mean,'; Interval=',C_interval_mean[1]-C_interval_mean[0])
print('90% CI of std:',C_interval_std,'; Interval=',C_interval_std[1]-C_interval_std[0])
print('90% CI of med:',C_interval_med,'; Interval=',C_interval_med[1]-C_interval_med[0])
```

An object of density $\rho $ and unknown mass $m$ is calibrated against a reference weight of density $\rho_{ref}$ and mass $m_{ref}$ using a balance operated in

### ${m}=({s}+{m_{ref}})\frac{1-\frac{{\rho_{air}}}{{\rho_{ref}}}}{1-(\frac{{\rho_{air}}}{{\rho_{ref}}+{\epsilon}})}$

with s being the mass of a small weight of density $\rho_{ref}$ added to the reference weight to give a balance.

and uncertainty of the parameters is defined as an error model: 

$E_s \sim N(0,(0.1 \times 10^{-3})^2)$,

$E_{\rho_{air}} \sim U(-0.3 \times 10^{3},0.3 \times 10^{3})$,

$E_{{\rho}_{ref}} \sim U(-0.5 \times 10^{6},0.5 \times 10^{6})$,

**(d) Uncertainty propagation is performed using correlated input data.<br>
(d-1) How would that affect computation of uncertainty of the output using perturbation method?<br>
(d-2) How would that affect computation of uncertainty of the output using Monte Carlo method?<br>**
Reference:<br>
(The mathematical fundamental is reference from corresponding course scribing on GitHub)<br>
(Ainsworth, Ronald. "A MONTE CARLO ANALYSIS OF THE EFFECTS OF COVARIANCE ON PROPAGATED UNCERTAINTIES.")

**(d_1)** The perturbation method is a method for computing the mean and standard deviation of a response quantity, which is based on the Taylor series expansion of the model response around the mean value of the input parameters.<br>
The progress of deriving the formulas is shown as follows:<br>

The Taylor series expansion of the model $M(x)$ around $x_0$. Let us replace x0 with $E[x]$ and get $E[M(x)]$:

$$E[Y]= E[M(X)]\approx M(u_x) + \sum^M_{i=1}\frac{\partial M}{\partial x_i}|_{x=x_i} E[(X_i-u_{x_i})] + \frac{1}{2} \sum^M_{i=1}\sum^M_{j=1} \frac{\partial^2M}{\partial x_i \partial x_j}|_{x=u_x} E[(X_i-u_{x_i})(X_j-u_{x_j})] $$

We know that $E[(x-u_x)]=0$ and $E[(x_i-u_{x_i})(x_j-u_{x_j})]=C_{ij}$ is the covariance of $x_i$ and $x_j$, so:

$$E[Y]\approx M(u_x) + \frac{1}{2} \sum^M_{i=1}\sum^M_{j=1} C_{ij}\frac{\partial^2M}{\partial x_i \partial x_j}|_{x=u_x}$$

The variance of the response is computed as follows:

$$Var[Y] \approx \sum^M_{i=1}\sum^M_{j=1} C_{ij} \frac{\partial M}{\partial x_i }|_{x=u_x} \frac{\partial M}{\partial x_j }|_{x=u_x} $$

As we can see, the $C_{ij}$ is the covariance of $x_i$ and $x_j$. If the perturbation method is performed at a circumstance where the input data is correlated, the covariance between the input data will no longer be 0. That means the covariance term will be introduced into the perturbation method.<br>

**(d_2)** Monte Carlo methods rely on repeated random sampling to obtain numerical results. Their essential idea is using randomness to solve problems that might be deterministic in principle. <br>
In principle, Monte Carlo methods can be used to solve any problem having a probabilistic interpretation. By the law of large numbers, integrals described by the expected value of some random variable can be approximated by taking the empirical mean (a.k.a. the sample mean) of independent samples of the variable.<br>
Here, Monte Carlo method is used as a method used for computing the mean and standard deviation of a response quantity, which is based on sampling from the original distribution. Assume that we get a sample set of $x$, then we can get:

$$Y = f(X_1,X_2,...,X_N)$$

$Y$ is the observed output and the $X_N$ are inputs that may or may not be observed. Solutions to the first-order Taylor series can be obtained directly by numerical methods, or by a method recently described using Lagrange polynomials. However it is also possible using matrix methods to solve for the propagation of uncertainty curves directly without the use of a first-order Taylor series approximation.<br>
If the argument of $Y = f(X_1,X_2,...,X_N)$ is rewritten as the vector $X$:<br>

$$Y=Xb$$

w the vector b represents the calibration coefficients and the vector X represents the calibration data. To predict what the interpolated mean response $\hat{Y}$ would be to some other set of inputs $X_h$, then the relationship as follows:<br>

$$\hat Y={X_h}b$$

If $Y$ is a random vector with expected variance-covariance matrix:<br>

$$\sigma^2\{Y\}=E\{[Y-E\{Y\}][Y-E\{Y\}]^T\}$$

By combining this result with previous wquation we obtain:<br>

$$\sigma^2\{\hat{Y}\}=E\{{X_h}X^{-1}\sigma^2\{Y\}({X_h}X^{-1})^T\}$$

The off-diagonal elements are the covariance and their estimation requires knowledge about the degree of correlation between the uncertainties along the diagonal of the matrix. The covariance is defined: <br>

$$\sigma(x_i,x_j)=\sigma(x_i,)\sigma(x_j,)r(x_i,x_j)$$

where $r(x_i,x_j)$ is the correlation coefficient. Hence when the input data is correlated, the Monte Carlo computation would also get involved with the covariance matrix of the input data.<br>

**(e) List MCMC methods that we studied and analyze their advantages and disadvantages <br>**
<br>

**Metropolis-Hastings:** <br>
In statistics, the Metropolis–Hastings algorithm is a MCMC method for obtaining a sequence of random samples from a probability distribution for which direct sampling is difficult. The original algorithm used in simulated annealing and MCMC’s is due to Metropolis. Later it got developped as it is not necessary to use a symmetric proposal distribution, and proposed that the proposed new state can be generated from any $q(y|x)$. Of course, the speed with which we reach the equilibrium distribution will depend on the choice of the proposal function.<br>
Algorithm:<br>
Firstly,initialize $x_0$<br>
then, for $i=0$ to $N-1$: $u~U(0,1)$, $x_*~q((x_*)|(x_(i)))$<br>
$if$ $$ u\leq\alpha(x_*)=min(1,\frac{\pi(x_*)q(x|x_*)}{\pi(x)q(x_*|x)}) $$
$$ x_{(i+1)}=x_* $$
$else$ $$     x_{(i+1)}=x_{(i)}$$
Advantages:<br>
•It can sample from any probability distributions, as long as the function $f(x)$ is propotional to density $P(x)$.<br>
Disadvantages:<br>
• M-H has the same drawbacks of MCMC, which related to the autocorrelation of the sample.<br>

**Gibbs Sampling:**<br>
In statistics, Gibbs sampling is a MCMC algorithm for obtaining a sequence of observations which are approximated from a specified multivariate probability distribution, when direct sampling is difficult. Gibbs sampling can be seen as a special case of the Metropolis-Hastings algorithm. <br>
Algorithm:<br>
Firstly, initialize $\{z_i:i=1,2,3,...,M\}$<br>
then, for $\tau=1,2,...,T$<br>
$sample$ $$z_1^{\tau+1}~p(z_1|z_2^{\tau},z_3^{\tau},...,z_M^{\tau})$$
$sample$ $$z_2^{\tau+1}~p(z_2|z_1^{\tau+1},z_3^{\tau},...,z_M^{\tau})$$
$$...... $$
$sample$ $$z_j^{\tau+1}~p(z_j|z_1^{\tau+1},...,z_{j-1}^{\tau+1},z_{j+1}^{\tau},...,z_M^{\tau})$$
$$...... $$
$sample$ $$z_M^{\tau+1}~p(z_M|z_1^{\tau+1},z_2^{\tau+1},...,z_{M-1}^{\tau+1})$$
Adavantages:<br>
• Proposals are always accepted<br>
• No tuning is required to proposal distribution<br>
Disadvantages:<br>
• Need to be able to draw random samples from the derived conditional probability distributions<br>
• Need to be able to derive conditional probability distributions<br>
• If parameters are correlated, it can be very slow<br>

**(f) Consider the model $y=x_1 \times x_2 + x_3 \times x_4 $ where $x_i \sim N(0, \sigma_i)$ for $i=1,...,4$. Find Sobol indexes $S_i$ and $S_{ij}$**<br>
The product of two independent Gaussians is given by (From Assignment 2):

$$\begin{aligned}\mu &=\mu_1 * \mu_2\\
\sigma^2 &= \sigma^2_1 * \sigma^2_2 + \sigma^2_1 * \mu^2_2 + \sigma^2_2 * \mu^2_1 
\end{aligned}$$

The sum of two Gaussians is given by (From Assignment 2):

$$\begin{gathered}\mu = \mu_1 + \mu_2 \\
\sigma^2 = \sigma^2_1 + \sigma^2_2
\end{gathered}$$

Therefore, for this problem:

$$var(y) = var(x_1 \times x_2) + var(x_3 \times x_4) $$

where

$$var(x_1 \times x_2)=\sigma^2_{x_1} * \sigma^2_{x_2} + \sigma^2_{x_1} * \mu^2_{x_2} + \sigma^2_{x_2} * \mu^2_{x1}=\sigma^2_{1} * \sigma^2_{2} $$ 
$$ $$
$$var(x_3 \times x_4)=\sigma^2_{x_3} * \sigma^2_{x_4} + \sigma^2_{x_3} * \mu^2_{x_4} + \sigma^2_{x_4} * \mu^2_{x_3}=\sigma^2_{3} * \sigma^2_{4} $$ 

then, 

$$var(y) = var(x_1 \times x_2) + var(x_3 \times x_4)=\sigma^2_{1}\cdot\sigma^2_{2}+\sigma^2_{3}\cdot\sigma^2_{4} $$

And the Sobol indexes $S_i$ can be written as:

$$S_1 = \frac{D_1}{D} = \frac{\sigma^2_{1}}{\sigma^2_{1}\cdot\sigma^2_{2}+\sigma^2_{3}\cdot\sigma^2_{4}}$$

$$S_2 = \frac{D_2}{D} = \frac{\sigma^2_{2}}{\sigma^2_{1}\cdot\sigma^2_{2}+\sigma^2_{3}\cdot\sigma^2_{4}}$$

$$S_3 = \frac{D_3}{D} = \frac{\sigma^2_{3}}{\sigma^2_{1}\cdot\sigma^2_{2}+\sigma^2_{3}\cdot\sigma^2_{4}}$$

$$S_4 = \frac{D_4}{D} = \frac{\sigma^2_{4}}{\sigma^2_{1}\cdot\sigma^2_{2}+\sigma^2_{3}\cdot\sigma^2_{4}}$$
<br>
which corresponds to the variance of the average $Y$ over all possible points of $X_i$, and it can be also explained as the sensitivity.

**(a)** Use pandas to load the data from the .csv file and plot the data:

```python
data = pd.read_csv('GP_Data.csv')

data=data.T
X = np.linspace(0,50,num=501).T
data = np.array(data)
plt.plot(X,data)
```

**(b)** Fit the data using Gausssian Processes with Squared Exponential Kernel. Plot the fitting with confidence intervals. 

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.gaussian_process.kernels import Matern, ExpSineSquared
np.random.seed(1)

X = np.atleast_2d(X.T)
y = data.ravel()


kernel = ExpSineSquared(length_scale=0.2, periodicity=1.0,
                                length_scale_bounds=(0.01, 100.0),
                                periodicity_bounds=(0.1, 10.0))
#kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=100)
gp.fit(X, y)
```

```python
x = np.atleast_2d(np.linspace(0, 50, 100)).T
y_pred, sigma = gp.predict(x, return_std=True)

plt.figure(figsize=(12,10))
plt.plot(X, y, 'r.', label=u'Observations')
plt.plot(x, y_pred, 'b-', linewidth=0.1,label=u'Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                        (y_pred + 1.9600 * sigma)[::-1]]),
         alpha=.9, fc='b', ec='None', label='95% confidence interval')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend(loc='upper left')
plt.show()
```

As we can see, the guassian processing with Squared Exponential Kernel allocated almost with the data, but the performance seems not that good enough, and the prediction vibrated between the observation curve, like non-convergence.

**(c)** Fit data using the Kernel of your choice.Plot the fitting with confidence intervals. In this case I choose to perform the Gaussian process with RBF kernel.

```python
kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=100)
gp.fit(X, y)
```

```python
x = np.atleast_2d(np.linspace(0, 50, 100)).T
y_pred, sigma = gp.predict(x, return_std=True)

plt.figure(figsize=(12,10))
plt.plot(X, y, 'r.', label=u'Observations')
plt.plot(x, y_pred, 'b-', linewidth=0.1,label=u'Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                        (y_pred + 1.9600 * sigma)[::-1]]),
         alpha=.9, fc='b', ec='None', label='95% confidence interval')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend(loc='upper left')
plt.show()
```

As we can see, the guassian processing with RBF Kernel allocated perfect with the observed data, and the performance seems good but seems a little over-fit. The prediction follows the data trend very well.
