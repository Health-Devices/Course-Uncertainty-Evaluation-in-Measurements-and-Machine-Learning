## Introduction

The basic operation of particle filters can be described as follows: Particle
filters use Monte Carlo simulation techniques to perform recursive Bayesian
estimation of the state of a system described by a general state space model.
The posterior pdf $p({x}_{0:n}|{y}_{0:n})$ is represented/approximated by a set of $N$
samples (particles) in the state space, and a set of associated probability
weights. This pointwise representation of the posterior can then be used to
answer statistical questions about the state.

More specifically: The particles $\{{x}^{(i)}\}_{i=1}^N$ are a Monte Carlo sample drawn
from an importance distribution, $q({x}_{0:n}|{y}_{0:n})$ and the weights are the corresponding
importance weights:
$$W_n^{(i)}\propto \frac{p({x}_{0:n}^{(i)}|{y}_{0:n})}{q({x}_{0:n}^{(i)}|{y}_{0:n})}$$

The Bayesian estimate of the expectation of the some function $f({x}_n)$ of the state, can be approximated using the particles and associated normalized importance weights as:
$$\mathcal{E}\{f({x}_n)\}\approx \sum_{i=1}^{N}W_n^{(i)}f({x}_n^{(i)})$$

In order to carry out the estimation recursively and online, the importance distribution evolves and is constructed sequentially using SIS.  Therefore it is assumed that the importance distribution factors as
$$q({x}_{0:n}|{y}_{0:n})\triangleq q({x}_n|{x}_{0:n-1},{y}_{0:n})q({x}_{0:n-1}|{y}_{0:n-1})$$  
enabling the weights to be updated recursively using the SIS weight update.  The particles and the corresponding estimate of the posterior pdf evolve with time by propagating the particles through the (known) state space model.  Enforcing the Markov structure assumed in recursive Bayesian estimation, the importance function becomes a function of only the immediate past 
$q({x}_n|{x}_{0:n-1},{y}_{0:n})=q({x}_n|{x}_{n-1},{y}_{n})$.  In this case the pdf represented by the particles is the filtered density $p({x}_n|{y}_{0:n})$.  Substituting the recursive Bayesian form of the filtered posterior into the SIS importance weight gives rise to the simplified SIS particle filter weight update:

$$W_n^{(i)} = W_{n-1}^{(i)}\frac{p({y}_n|{x}_n^{(i)})p({x}_n^{(i)}|{x}_{n-1}^{(i)})}{q({x}_n^{(i)}|{x}_{n-1}^{(i)},{y}_n)}$$ (1)

Drawing particles and updating the weights form the basis of particle filtering algorithms. A problem with sequential importance sampling is that the variance of the weights updated according to (equation 1) can only increase over time.  Consequently, after a few iterations, only one particle will have a non-zero normalized weight, leading to poor performance of the particle filter and wasted computational effort updating the zero-weight particles. Resampling is used to combat this problem. A measure of the degeneracy of the algorithm, which can be used to decide when resampling is necessary, is given by the effective sample size, $N_{eff}$.  In theory $N_{eff}$ is the number of samples scaled by the variance of the weights, in practice it is estimated by:

$$\hat{N}_{eff} = \frac{1}{\sum_{i=1}^{N}(W_n^{(i)})^2}$$ (2)

and if the $N_{eff}$ falls below some threshold $N_T$, resampling is performed.  

The combination of sampling, importance weighting and resampling steps together form the generic particle filtering algorithm.  This algorithm, the SIS particle filter, is summarized.

By eliminating zero or low weight particles and replicating higher weight particles, resampling side-steps the wasted computations imposed by the weight degeneracy problem.  However, resampling does not solve the problem of degrading accuracy caused by sample impoverishment. Since particles with high weights are selected many times during the resampling step, the resulting particle set is highly correlated and exhibits low diversity.  This set of particle may not adequately represent the true distribution, reducing the accuracy of the estimate.  According to [1], in general the accuracy of an estimate of a function of the posterior, $\hat{f}({x}_n)$, can only decrease after resampling, as resampling introduces extra Monte Carlo variation.

### SIR Particle filter [2]
The SIR particle filter is one of the earliest and simplest particle filter algorithms.  Executing the algorithm requires: that the state space model be known; that the process noise ${d}_n$, and the prior $p_x({x}_0)$ can be sampled from; and that the liklihood $p({y}_n|{x}_n)$ can be evaluated pointwise, up to a proportionality.  The importance density $q({x}_n|{x}_{n-1}^{(i)},{y}_{n})$ used by the SIR particle filter is the transitional prior $p({x}_n|{x}_{n-1}^{(i)})$.  This importance density is convenient as sampling from it only requires sampling from the process noise and propagating the particles and sampled noise through the state space transition function.  In the SIR particle filter algorithm resampling is performed at every time index, so the normalized weights are always $1/N$ and don't need to be propagated.  This also simplifies the evaluation of the un-normalized particle weights. Since the importance function cancels the transition prior in(equation 1), and the weights carried forward are all equal, the weight update simplifies to: 
$$\tilde{W}_n^{(i)}=p({y}_n|{x}_n^{(i)})$$  (3)

A main advantage of the SIR filter is its simplicity, both in terms of evaluating the importance weights and sampling the particles.  The most obvious drawback is the fact that the importance density does not take into account the current measurement.  Knowledge of the measurement can encourage particles to be drawn from higher probability regions of the state space.  Not using the measurement reduces the efficiency of the SIR filter and makes it more sensitive to outliers.  In addition, while resampling at every step reduces the memory requirements, it can also lead to a rapid reduction in particle diversity.

#### ALGORITHM: Sequential Importance Sampling Particle Filter

[1] **for** {$n \geq 0$} **do**

[2] **for** {$i=1,...,N$}

$$tilde{W}_n^{(i)} \gets W_{n-1}^{(i)}\frac{p({y}_n|{x}_n^{(i)})p({x}_n^{(i)}|{x}_{n-1}^{(i)})}{q({x}_n^{(i)}|{x}_{n-1}^{(i)},{y}_n)}$$

[5] **end for**

[6] **for** {$i=1,...,N$} **do**

[7] Normalize the importance weights.
$$W_n^{(i)} \gets \frac{\tilde{W}_n^{(i)}}{\sum_{j=1}^{N}\tilde{W}_n^{(j)}}$$

[8] **end for**

[9] Compute Bayesian posterior estimate.
$$ \hat{f}({x}_n)\gets \sum_{i=1}^{N}W_n^{(i)}f({x}_n^{(i)})$$

[10] Estimate $N_{eff}$ 

$$\hat{N}_{eff} \gets \frac{1}{\sum_{i=1}^{N}(W_n^{(i)})^2}$$

[11] **if** $N_{eff}<N_T$ **then**
    
[12] Resample $\{{x}_{n+1|n}^{(i)},W_{n+1|n}^{(i)}\}_1^N\gets\{{x}_n^{(i)*},1/N\}_1^N$

[13] **end if**

[14] **end for**


#### ALGORITHM  Sequential Importance Resampling Particle Filter

*Initialization*

[1] **for** {$i=1,...,N$} **do**

[2]Draw ${x}_n^{(i)}\sim p_x({x}_0)$

[3]**end for**

*Iteration*

[4]**for** {$n \geq 1$} **do**

[5]**for**{$i=1,...,N$} **do**

[6]Draw samples ${d}_{n-1}^{(i)}\sim p_d({d}_{n-1})$.

[7]Propagate the particles and the sampled noise through the state transition function
$${x}_n^{(i)}\gets {f}_{n-1}({x}_{n-1}^{(i)},{d}_{n-1}^{(i)})$$

[8] Evaluate the importance weights, up to a constant
$$\tilde{W}_n^{(i)} \gets p({y}_n|{x}_n^{(i)})$$

[9]**end for**

[11] Normalize the importance weights.

$$W_n^{(i)} \gets \frac{\tilde{W}_n^{(i)}}{\sum_{j=1}^{N}\tilde{W}_n^{(j)}}$$

[12]**end for**

```python
import numpy as np
from numpy import random, zeros
import matplotlib.pyplot as plt
from math import sqrt, cos
nIter=50
N=500
q=sqrt(10)
r=1
```

```python
w=q*random.randn(1,nIter)
v=random.randn(1,nIter)
x=zeros((nIter,1))
x[0]=0.1
z=zeros((nIter,1))
```

```python
for k in range(1,nIter):

    x[k]=0.5*x[k-1]+25*x[k-1]/(1+x[k-1]*x[k-1])+8*cos(1.2*(k-1))+w[0][k]
    z[k]=(x[k]*x[k])/20+v[0][k]
```

```python
xp=sqrt(2)*random.randn(1,N)
xh=zeros((nIter,1))
v_xh=zeros((nIter,1))
```

```python
def resampling(weights):
    N=len(weights)
    u=random.rand()/float(N)
    s=0
    idx=np.array([])
    for n in range(0,N):
        k=0
        s=s+weights[0][n]
        while s > u:
            k = k+1
            u = u+1/float(N)
            idx=np.appendix(idx,n*np.ones(1,k))
```

```python
for k in range(1, nIter):
    #prediction step
    w=q*random.randn(1,N)
    #propagate particles through system
    xp=0.5*xp + np.divide(25*xp,(1+xp*xp))+8*cos(1.2*(k-1))+w
    #compute weights
    weights= np.exp(-(z[k]-xp*xp/20)/2)
    #normalize weights
    if not np.any(weights):
        weights[:] = np.finfo(float).eps
    weights=weights/np.sum(weights)
    
        
    if np.any(np.isnan(weights[:])|np.isinf(weights[:])):
        raise ValueError('Weight problem')
        
    #compute posterior estimates
    xh[k] = np.sum(np.multiply(weights,xp))
    
    #estimate variance of estimate
    v_xh[k] = np.var(xp)
    
    #plot prior pdf
    Nx,X = np.histogram(xp, 50)
    y_pos = np.arange(len(Nx))
    fig, ax=plt.subplots()
    ax.bar(X[:-1], Nx,width=np.diff(X), ec="k", align="edge")
    plt.show()
    #plt.hold(True)
    
    #Resample
    idx = resampling(weights)
    xp = xp[idx][0]
    
    #plot posterior pdf
    Nx,X = np.histogram(xp, 50)
    y_pos = np.arange(len(Nx))
    fig, ax = plt.subplots()
    ax.bar(X[:-1], Nx, width=np.diff(X), ec="k", align="edge")
    plt.show()
```

### Auxiliary Particle Filter

One weakness with the standard SIR particle filter is that the particle draws
are taken “blind”, that is the current measurement is not used. Furthermore,when the predicted state lies in the tail of the likelihood there will be very few samples with a high likelihood leading to rapid particle diversity depletion; ie. the algorithm is sensitive to outliers. The auxiliary SIR (ASIR) particle filter attempts to multiply particles with high predictive likelihoods (high $p({y}_n| {x}^{(i)}_{0:n−1})$) prior to the particle propagation step. This increases the
number and diversity of particles in the areas of high likelihood. In order to achieve this goal, the auxiliary particle filter switches the order of sampling and resampling used by SIR: When the new measurement is available, the particles are resampled using some measure of the predictive likelihood as the weighting function. These high predicitive likelihood resampled particles are then propagated through the transition and measurement functions as in the SIS and SIR particle filters.
The justification of the algorithm can be examined mathematically by
introducing an auxiliary variable $\xi$ $\epsilon$ {1, . . . ,N}. Draws are taken from the joint importance density $q(x_n,\xi|y_{0:n})$, producing sample pairs {$x^{(i)}_n ,\xi^{(i)}$} where $\xi^{(i)}$ is the index of particle $x^{(i)}_n$ at time n − 1. The auxiliary variable $\xi^{(i)}$ determines the parent particle of $x^{(i)}_n$ . Using Bayes rule, the joint density $p(x_n, \xi |y_{0:n})$ is given by:

$$ p({x}_n,\xi=i|{y}_{0:n}) \propto p({y}_n|{x}_n)p({x}_n,\xi=i|{y}_{0:n-1})$$

$$=p({y}_n|{x}_n)p({x}_n|\xi=i,{y}_{0:n-1})p(\xi=i|{y}_{0:n-1})$$

$$=p({y}_n|{x}_n)p({x}_n|{x}_{n-1}^{(i)})W_{n-1}^{(i)}$$

The last line follows from the fact that $p(\xi=i|{y}_{0:n-1})$ expresses the probability that particle ${x}_{n-1}^{(i)}$ will be resampled, and is governed by the particle weights.  $p({x}_n|\xi=i,{y}_{0:n-1})$ gives the transition probability for a given parent particle, and is dictated by the Markov model. 

The operation of the ASIR filter is a two-step procedure.  In the first step a set of ``parent particles'' are chosen, ie. a set of indices $\{\xi^{(i)}\}_1^N$ are simulated with probability $p(\xi|{y}_{0:n})$; in the second step the particles $\{{x}_{n}^{(i)}\}_1^N$ are simulated by propagating the selected parent particles.  This can be seen as drawing a sample pair $({x}_n,\xi)$ from $p({x}_n,\xi|{y}_{0:n})$, then dropping the auxiliary variable $\xi$ to obtain a draw from the marginal density $p({x}_n|{y}_{0:n})$.  The proposal distribution used to draw the samples pairs is chosen to factor as:

$$q({x}_n,\xi|{y}_{0:n}) \propto q(\xi|{y}_{0:n})q({x}_n|\xi,{y}_{0:n})$$

Where

$$q(\xi^{(i)}|{y}_{0:n}) \propto p({y}_n|\mu_n^{(i)})W_{n-1}^{(i)}$$

$$q({x}_n|\xi^{(i)},{y}_{0:n}) \triangleq p({x}_n|{x}_{n-1}^{(i)})$$


The random variable $\mu_n^{(i)}$ is some statistic that characterizes $p({x}_n|{x}_{n-1})$.  A popular choice is the mean $\mu_{n}^{(i)}=\mathcal{E}\{{x}_n|{x}_{n-1}^{(i)}\}={f}_{n-1}({x}_{n-1}^{(i)},\mathcal{E}\{d_{n-1}\})$; which is obtained by propagating the previous samples through the state transition equation with the mean process noise.  The set of parent particles are selected by resampling from the set of particles using the predictive likelihood as the resampling weight.  The predictive likelihood is estimated using $\mu_n^{(i)}$, generating a set of resampling indices $\{\xi^{(i)}\}$ where $p(\xi|{y}_{0:n})\propto p({y}_n|\mu_n^{(i)})W_{n-1}^{(i)}$ as required.

The weight update step is then given by:

$$W_n^{(i)}=W_{n-1}^{(\xi=i)}\frac{p({y}_n|{x}_n^{(i)})p({x}_n^{(i)}|{x}_n^{(\xi=i)})}{q({x}_n^{(i)},\xi^{(i)}|{y}_{0:n})}$$
$$\propto \frac{p({y}_n|{x}_n^{(i)})}{p({y}_n|\mu_n^{(\xi=i)})}$$

$$W_n^{(i)}=W_{n-1}^{(\xi=i)}\frac{p({y}_n|{x}_n^{(i)})p({x}_n^{(i)}|{x}_n^{(\xi=i)})}{q({x}_n^{(i)},\xi^{(i)}|{y}_{0:n})}$$
$$\propto \frac{p({y}_n|{x}_n^{(i)})}{p({y}_n|\mu_n^{(\xi=i)})}$$ (4)

The ASIR particle filtering algorithm is summarized in the algorithm below.  For illustrative purposes, the mean is chosen to characterize the relationship between a particle and its parent, ie. $\mu_n^{(i)}=\mathcal{E}\{p({x}_n|{x}_{n-1})\}$.  When the process noise is small, $p({x}_{n}|{x}_{n-1}^{(i)})$ is well described by $\mu_n^{(i)}$ and the ASIR filter is more robust to outliers, and can achieve better performance than SIR.  If the process noise is large, the transitional probability is not well represented by the single point $\mu_n^{(i)}$, and ASIR can perform worse than SIR.

#### ALGORITHM  Auxiliary Sequential Importance Resampling Particle Filter
*Initialization*

[1]**for** {$i=1,...,N$} **do**

[2]Draw ${x}_n^{(i)}\sim p_x({x}_0)$

[3]end for

*Iteration*

[4]**for**{$n \geq 1$} **do**

[5]**for**{$i=1,...,N$} **do**

[6] $\mu_n^{(i)}\gets {f}_{n-1}({x}_{n-1}^{(i)},\mathcal{E}\{d_{n-1}\})$

[7]Calculate the first stage importance weights up to a constant.
        $$\tilde{W}_n^{(i)}\gets W_{n-1}^{(i)}p({y}_n|\mu_{n}^{(i)})$$
[8]**end for**

[9]Normalize the first stage weights.

[10]Resample using the first stage weights, to obtain $\{{x}_{n-1}^{(i)},\xi_n^{(i)}\}_1^N$.

[11]**for**{$i=1,..,N$} **do**

[12]Draw samples ${d}_{n-1}^{(i)}\sim p_d({d}_{n-1})$.
 
[13]Propagate the particles and the sampled noise through the state transition function
$${x}_n^{(i)}\gets {f}_{n-1}({x}_{n-1}^{(i)},{d}_{n-1}^{(i)})$$

[14]Evaluate the importance weights, up to a constant

$$\tilde{W}_n^{(i)} \gets \frac{p({y}_n|{x}_n^{(i)})}{p({y}_n|\mu_n^{(\xi=i)})}$$

[15]**end for**

[16]Normalize the importance weights and compute Bayesian posterior estimate.
$$\hat{f}({x}_n)\gets \sum_{i=1}^{N}W_n^{(i)}f({x}_n^{(i)})$$

[17]**end for**

### Rao-Blackwellization of Particle Filters [3][4]

When the dimension of the state space is high, many particles are required to achieve good performance.  Rao-Blackwellization is a technique that exploits the structure of the state-space model to reduce the dimension of the sampling space, improving sampling efficiency and achieving a lower variance estimate for a given number of particles.  The general idea behind Rao-Blackwellization is as follows: Consider the case where the state vector can be partitioned into two sub-states ${x}_{0:n}=[{r}_{0:n},{b}_{0:n}]$.  The chain rule of probability allows the joint posterior to be decomposed as 
$$p({r}_{0:n},{b}_{0:n}|{y}_{0:n})=p({b}_{0:n}|{r}_{0:n},{y}_{0:n})p({r}_{0:n}|{y}_{0:n})$$
If $p({b}_{0:n}|{r}_{0:n},{y}_{0:n})$ can be computed analytically (eg. with a Kalman or HMM filter), then Monte Carlo inference need only be carried out to estimate $p({r}_{0:n}|{y}_{0:n})$, which is of a lower dimension than $p({x}_{0:n}|{y}_{0:n})$.  The result is that for a given number of particles, the posterior estimate $p({r}_{0:n},{b}_{0:n}|{y}_{0:n})=p({x}_{0:n}|{y}_{0:n})$ has a lower variance. The improved estimate does not come without a cost: analytically computing $p({b}_{0:n}|{r}_{0:n},{y}_{0:n})$ requires additional computation, which may not be insignificant.  The concept of Rao-Blackwellization is explained most easily in the context of an application.

#### Rao-Blackwellized particle filters for speech enhancement [5] [6]

In RBPF speech enhancement, the speech signal at time $n$, $x_n$ is assumed to be corrupted by additive measurement noise $v_n$ giving the measurement signal $z_n$:
$$z_n = x_n + v_n$$ (5)

The clean speech signal is assumed to follow a $p^{th}$ order time-varying autoregressive (TVAR) model:
$$x_{n} = \sum_{i=1}^p a_{n}(i)x_{n-i} + d_n$$ (6)
$$= {a}_n^T{x}_n + d_n$$ (7)
where: 
${x}_n = [x_{n-1} \ldots x_{n-p}]^T $(8)
is the signal vector,
${a}_n=[a_n(1)\ldots a_n(p)]^T $ (9)
is the vector of autoregression coefficients at time $n$, and $d_n$ is the excitation sequence.  

Defining the state-transition and measurement matrices ${F}_n$ and ${G}$, (equation 6) can be described by the following linear state-space model:
$${F}_n  = \left[\begin{array}{cc}{{a}_{n}^T}\\{I}_{p-1}   {0}_{p-1\times1}\end{array}\right]$$ (10)
$${G} = \left[\begin{array}{c}1\\{0}_{p-1\times 1}\end{array}\right]$$(11)
$${x}_n = {F}_n{x}_{n-1} + \left[\begin{array}{c}d_n\\{0}_{p-1\times 1}\end{array}\right]$$ (12)
$$z_n = {G}{x}_n + v_n$$(13)

where (equation 12) and (equation 13) are the state-space transition and measurement equations respectively.

If all of the TVAR parameters are known, and $d_n\sim\mathcal{N}(0,\sigma_d^2)$ and $v_n\sim\mathcal{N}(0,\sigma_v^2)$ are uncorrelated zero-mean white Gaussian noise processes, then an MMSE optimal estimate of $x_n$ can be estimated using a Kalman filter.  Non-white disturbances can be accommodated by augmenting the state-vector so that it contains two AR models, one each for the speech and noise components.  An alternative to measuring the TVAR parameters directly from the noisy signal is to estimate them using Monte Carlo methods, this is the approach taken in Rao-Blackwellized particle filter speech enhancement. [6][5] 

In the RBPF speech enhancement algorithm of [5],the TVAR parameters themselves are assumed to evolve stochastically.  A constrained Gaussian random walk model is assumed for the AR coefficient vector, where the roots of the selected vector are forced to lie inside the unit circle (a sufficient but not necessary condition for stability of the TVAR model):

$$ p({a}_n|{a}_{n-1}) \propto \begin{cases}\mathcal{N}({a}_n|{a}_{n-1},\Delta_a^2{I}) \text{If stable}\\0 \text{Otherwise,}\end{cases}$$ (14)

where $\mathcal{N}(x|\mu,\sigma^2)$ denotes a Gaussian pdf with mean $\mu$ and (co-)variance $\sigma^2$ evaluated at $x$.  The speech and noise variances are assumed to evolve according a random walk in the log-domain, to ensure that the variances remain positive:  
$$p(\log(\sigma_{d_n}^2)|\log(\sigma_{d_{n-1}}^2))=\mathcal{N}(\log(\sigma_{d_n}^2),\delta_d^2)$$ (15)
$$ p(\log(\sigma_{v_n}^2)|\log(\sigma_{v_{n-1}}^2))=\mathcal{N}(\log(\sigma_{v_n}^2),\delta_v^2)$$ (16)

The hyper-parameters $(\Delta_a^2,\delta_d^2,\delta_v^2)$ define the variances of the random walks taken by the TVAR parameters.

Defining the vector of TVAR parameters:
$${\theta}_n = [{a}_n^T \sigma_{d_n}^2 \sigma_{v_n}^2]^T$$ (17)

we have an augmented state vector $[{x}_n^T,{\theta}^T]^T$.  The transition equations describing the evolution of this augmented state vector are stochastic and not linear, however, conditioned on ${\theta}_n$, ${x}_n$ evolves according to a linear Gaussian state space model as described by (equation 12) and (equation 13), and the MMSE optimal estimate of ${x}_n$ can be obtained using the Kalman filter equations.  

A RBPF is essentially two interacting algorithms: the particle filter generates $N$ candidate AR parameter vectors $\{{\theta_n}^{(i)}\}_{i=1}^N$ based on a model of the parameter evolution.  The estimated parameters are then supplied to a Kalman filter, which enhances speech and provides an error covariance estimate.  The particle filter then weights the contributions of the candidate enhanced speech vectors using the conditional likelihood of the measurement signal given the mean and error covariance estimates provided by the Kalman filter. The candidate AR vectors and their corresponding Kalman mean and covariance are then re-sampled $N$ times with replacement to form the *a-posteriori* set of candidates for the next iteration.
The RBPF speech enhancement algorithm is summarized in algorithm below.  Systematic re-sampling is carried out at each sampling interval to prevent particle degeneracy.

#### ALGORITHM Rao-Blackwellized Particle Filtering

[1]**for** $n \geq 0$} **do**

[2] **for** $i=1,...,N$} **do*

[3]Draw candidate TVAR parameters ${\theta}_n^{(i)}\sim p({\theta}_n|\theta_{n-1}^{(i)})$.

[4]Obtain ${F}_n^{(i)},\sigma_{d_n}^{2,(i)},\sigma_{v_n}^{2,(i)}$ from ${\theta}_n^{(i)}$.

[5]Use the TVAR parameters to update the Kalman statistics:

$${x}_{n|n-1}^{(i)} = {F}_n^{(i)}{x}_n^{(i)}$$

$${P}_{n|n-1}^{(i)} = {F}_n^{(i)}{P}_{n-1|n-1}^{(i)}{F}_n^{T,(i)} + \sigma_{d_n}^{2,(i)}$$

$${T}_n^{(i)} = {G}{P}_{n|n-1}^{(i)}{G}^{T} + \sigma_{v_n}^{2,(i)}$$

$${K} = {P}_{n|n-1}{G}^T{T}_n^{-1,(i)}$$

$${x}_{n|n}^{(i)} = {x}_{n}^{(i)} + {K}(z_n - {G}{x}_{n|n-1}^{(i)})$$

$${P}_{n|n}^{(i)} = ({I} - {K}{G}){P}_{n|n-1}^{(i)}$$

[6]Compute the un-normalized importance weights:
$$\tilde{W}_n^{(i)} = \mathcal{N}(z_n|{G}{x}_{n|n-1}^{(i)},{T}_n^{(i)})$$

[7]end for

[8]Normalise the importance weights.
$$\{{W}_n\}_1^N \gets \frac{\{{\tilde{W}}_n\}_1^N}{\sum_{i=1}^{N}\tilde{W}_n^{(i)}}$$

[9]Compute Bayesian posterior estimate.
$$\hat{x}_n\gets \sum_{i=1}^{N}W_n^{(i)}{x}_n^{(i)}$$

[10]Resample: $\{{x}^{(i)},{P}^{(i)},{\theta}^{(i)}\}$

[11]**end for**
