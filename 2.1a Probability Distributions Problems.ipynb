$\textbf{Problem 1}$<br>
The joint density function for the random variables $(X,Y)$ is given by :
\begin{align}
f(x,y)&=10xy^2,\,0<x<y<1\\
&=0 \,\text{elsewhere}
\end{align}
(a) Find the marginal distribution of $X$ and $Y$<br>
(b)Compute the probability of $0<x<\frac{1}{2}$, $\frac{1}{4}<y<\frac{1}{2}$

$\textbf{Answers of (a):}$<br>
The marginal distribution of $X$:
\begin{align}
f_X(x) = \displaystyle{\int_0^1}f(x,y) dy = \int_{D_{y}}10 x y^2 dy= \int_x^110 x y^2 dy = \frac{10}{3} x \left[y^3\right]_1^x
=\frac{10}{3} x (1-x^3)= \frac{10}{3}x - \frac{10}{3}x^4
\end{align} 
The marginal distribution of $Y$:
\begin{align}
f_Y(x) = \displaystyle{\int_0^1}f(x,y) dx = \int_{D_{x_1}} 10 x y^2 dx + \int_{D_{x_2}}10 x y^2 dx = \int_0^y10 x y^2 dx + \int_0^110 x y^2 dx = 5 y^2 \left[x^2\right]_0^y + 0
=5 y^2 (y^2 -0)=5y^4
\end{align} 


$\textbf{Answers of (b):}$<br>
The probability of $ 0<x<\frac{1}{2}$, $\frac{1}{4}<y<\frac{1}{2}:$ <br>
\begin{align}
P(0<x<\frac{1}{2},\, \frac{1}{4}<y<\frac{1}{2})=\iint_{R_1} 10xy^2 dy dx = \int_{\frac{1}{4}}^{\frac{1}{2}}\int_{0}^{\frac{1}{2}}10xy^2 dy dx \\
= \frac{10}{2}\int_{\frac{1}{4}}^{\frac{1}{2}}[x^2y^2]_{0}^{\frac{1}{2}}dy
=\frac{5}{4}\int_{\frac{1}{4}}^{\frac{1}{2}}y^2
=0.045
\end{align} 

$\textbf{Problem 2}$<br>
Consider a medical test for a disease. The test has a probability of 0.95 of correctly or positively detecting an
infected person (this is the sensitivity), while it has a probability of 0.90 of correcctly identifying a healthy person (this is called specificity). In the population, only 3% of the people have the disease.

(a) What is the probability that a person testing positive is actually infected?<br>
(b) What is the probability that a person testing negetive is actually infected?


$\textbf{Answer:}$<br>
Define $X$: detection result ($X=1$: prediction is positive; $X=0$: predition is negtive)<br>
Define $Y$: class ($Y=1$: infected class; $Y=0$: healthy class)<br>
$P(X=1|Y=1)=0.95$<br>
$P(X=0|Y=1)=0.05$<br>
$P(X=1|Y=0)=0.10$<br>
$P(Y=1)=0.03$<br>
$P(Y=0)=0.97$<br>
$P(X=1)=P(X=1|Y=1)P(Y=1)+P(X=1|Y=0)P(Y=0)=0.95*0.03+0.10*0.97=0.1255$<br>
$\textbf{(a):}$ The probability that a person testing positive is actually infected:<br>
$P(Y=1|X=1) = \frac{P(X=1,Y=1)}{P(X=1)} = \frac{P(X=1|Y=1)P(Y=1)}{P(X=1)} = \frac{0.95*0.03}{0.1255} = 0.227$<br>
$\textbf{(b):}$ The probability that a person testing negetive is actually infected:<br>
$P(Y=1|X=0) = \frac{P(X=0,Y=1)}{P(X=0)} = \frac{P(X=0|Y=1)P(Y=1)}{P(X=0)} = \frac{0.05*0.03}{1-0.1255} = 0.0017$

 Firstly, two variables $X$ and $Y$ were defined; from question, conditonal probabilities( $P(X=1|Y=1)=0.95$, $P(X=0|Y=1)=0.05$, $P(X=1|Y=0)=0.10$ ) and marginal probabilities $P(Y=0)=0.97$ and $P(Y=1)=0.03$ were obtained. Then $P(Y=1|X=1)$ and $P(Y=1|X=0)$ were calculated by using Bayes Theorem: P(x,y)=P(y|x)P(x)=P(x|y)P(y), we can get the answers.

$\textbf{Problem 3}$<br>
Sulfar Dioxide concentrations in the air samples taken in a certain region has been found to be well represented by lognormal distribution with parameters $\mu = 2.1$ and $\sigma = 0.8$.(Hint: convert the distribution into standard normal).<br>
(a) What proportions of air samples have concentration between 3 and 6?<br>
(b) What proportion do not exceed 10?<br>
(c) What interval (a,b) is such that 95% of all air samples have concentration values in this interval, with 2.5% have values less than a, and 2.5% have values exceeding b?<br>
Write Python code to simulate cases (a), (b) and (c)

```python
from scipy.stats import norm
pl=norm.cdf(np.log(3), loc=2.1, scale=0.8) # P(Y<log 3) in normal distribution
pr=norm.cdf(np.log(6), loc=2.1, scale=0.8) # P(Y<log 6) in normal distribution
prl=pr-pl # P(log 3 < Y < log 6)
print("Proportions of air samples have concentration between 3 and 6:", prl)

p10=norm.cdf(np.log(10), loc=2.1, scale=0.8) # P(Y<log 10) in normal distribution
print("Proportion do not exceed 10:{0}".format(p10))
loga,logb=norm.interval(0.95, loc=2.1, scale=0.8) # P(loga<Y<logb)= 0.95 in normal distribution
a=np.exp(loga) # change back to lognormal distribution
b=np.exp(logb) # change back to lognormal distribution
print("95% of all air samples have concentration values in [{0},{1}]".format(a,b))
```

    $Y = \ln X \sim \text{N}(\mu,\sigma^2)$, when $X = 3, 6, 10, \,Y = \ln3, \ln6, \ln10$. Function norm.cdf() was used to resove probabilities and function normal.interval() was used to calculate interval.

$\textbf{Problem 4}$<br>
A building can have structural damage (D) during its lifetime only from fire (F), high winds(W) and strong
earthquakes(E). Thus, F, W and E are collectively exhaustive events. Further assume that the building will not be
structurally damaged simultaneously by F, W and E, thus making them mutually exclusive events. The probabilities
of structural damage to the building if these events occur are estimated to be 0.005, 0.01 and 0.05 , respectively.
The probabilities of occurrence of F, W and E during the life of the building are 0.5, 0.3 and 0.2, respectively.<br>
(a) What is the probability that the building will suffer structural damage during its lifetime?<br>
(b) If the building has suffered structural damage, what is the probability that it was caused by F? By W? By E?

$\textbf{Answer:}$<br>
From question we have:<br>
F, W and E are mutually exclusive, so:<br>
$P(D|F) = 0.005$<br>
$P(D|W) = 0.01$<br>
$P(D|E) = 0.05$<br>
$P(F) = 0.5$<br>
$P(W) = 0.3$<br>
$P(E) = 0.2$<br>
$\textbf{(a):}$ The probability that the building will suffer structural damage during its lifetime:<br>
$P(D) = P(D|F)P(F) + P(D|W)P(W) + P(D|E)P(E) = 0.005*0.5 + 0.01*0.3 + 0.05*0.2 = 0.0155$<br>
$\textbf{(b):}$The probability of structural damage that was caused by F:<br>
$P(F|D) = \frac{P(D|F)P(F)}{P(D)} = \frac{0.005*0.5}{0.0155} = 0.1613$<br>
The probability of structural damage that was caused by W:<br>
$P(W|D) = \frac{P(D|W)P(W)}{P(D)} = \frac{0.01*0.3}{0.0155} = 0.1935$<br>
The probability of structural damage that was caused by E:<br>
$P(E|D) = \frac{P(D|E)P(E)}{P(D)} = \frac{0.05*0.2}{0.0155} = 0.6452$

Because F, W and E are mutually exclusive, The probability that the building will suffer structural damage during its lifetime
P(D) is sum of three probabilities of structural damages that were caused by F, W and E. Then using Bayes Theorem: P(x,y)=P(y|x)P(x)=P(x|y)P(y), the results were obtained.

$\textbf{Answer(a):}$<br><br>
$\textbf{Bernoulli Disribution}$<br><br>
Probability mass function: $p(x|p)=\text{B}(x|p)=p^x(1-p)^{1-x}$, $p\in(0,1)$, $x\in\{0,1\}$.
Sufficient statistics: $\text{E}[x]=p$, $\text{Var}[x]=p(1-p)$.

Important property: $p(x=1)=p$, $p(x=0)=1-p$.

Application: Model event having two states.<br><br>

$\textbf{Binomial Distribution}$<br><br>
Probability mass function: $p(x|n,p)=\text{BN}(x|n,p)=\binom{n}{x}p^x(1-p)^{n-x})$.

Sufficient statistics: $\text{E}[x]=np$, $\text{Var}[x]=np(1-p)$.

Important property: If $X\sim \text{BN}(n,p)$, $Y\sim \text{BN}(m,p)$, and $X \perp Y$, then $X+Y\sim \text{BN}(n+m,p)$.

Application: Model number of successes in $n$ trials with replacement.<br><br>

$\textbf{Multinoulli Distribution}$<br><br>
Probablity mass function: $p(x_1,\cdots,x_M|p_1,\cdots,p_M)=\text{M}(x_1,\cdots,X_M|p_1,\cdots,p_M)=p_1^{x_1} \cdots p_M^{x_M}$, $p_m\ge0$, $\sum_{m=1}^M p_m=1$, $x_m\in\{0,1\}$, $\sum_{m=1}^M x_m=1$.

Sufficient statistics: $\text{E}[x_m=1]=p_m$, $\text{Var}[x_m=1]=p_m(1-p_m)$.

Application: The generalization of Bernoulli distribution. Model event having more than two states.<br><br>

$\textbf{Multinomail Distribution}$<br><br>
Probability mass function: $p(x_1,\cdots,x_M|n,p_1,\cdots,p_M)=\text{MN}(x_1,\cdots,X_M|n,p_1,\cdots,p_M)=\frac{n!}{\prod_{m=1}^M x_m!} p_1^{x_1} \cdots p_M^{x_M}$, $p_m\ge0$, $\sum_{m=1}^M p_m=1$, $x_m\in\{0,1,\cdots,n\}$, $\sum_{m=1}^M x_m=n$.

Sufficient statistics: $\text{E}[x_m]=np_m$, $\text{Var}[x]=np_m(1-p_m)$.

Application: The generalization of binomial distribution. Model the number of successes of one out of $M$ categories in $n$ trials.<br><br>

$\textbf{Poisson Distribution}$<br><br>
Probability density function: $p(x|\lambda)= \text{P}(x|\lambda)=\frac{e^{-\lambda}}{x!} \lambda^x $, $\lambda>0$, $x\ge0$ 

Sufficient statistics: $\text{E}[x]=\lambda$, $\text{Var}[x]=\lambda=e^{\theta}$.

Important property: If $X_i \sim \text{P}(\lambda_i)$ for $i=1,...,n$, then $X=\sum_{i=1}^n X_i \sim \text{P}(\sum_{i=1}^n \lambda_i)$.

Application: Model the counts of an event within an interval of time or space where the event occurs at a fixed rate and the occurances are independent.<br><br>

$\textbf{Normal Distribution}$<br><br>
Probability density function: $p(x|\mu, \sigma^2) = \text{N}(x|\mu,\delta^2) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp(-\frac{(x-\mu)^2}{2\sigma^2})$.

Sufficient statistics: $\text{E}[x]=\mu$, $\text{Var}[x]=\sigma^2$.

Important property: $\frac{X-\mu}{\sigma}\sim N(0,1)$; If independent random variables $X_i\sim \text{N}(\mu_i,\sigma_i^2)$ for $i=1,\cdots,n$, then $\sum_{i=1}^n X_i \sim N(\sum_{i=1}^n \mu_i, \sum_{i=1}^n \sigma_i^2)$; If independent random variables $X_i\sim \text{N}(0,1)$ for $i=1,\cdots,n$, then $\sum_{i=1}^n X_i \sim \chi^2(n)$; and central limit theorem.

Application: Model real-valued signals.

$\textbf{Exponential Distribution}$<br><br>
Probability density function: $p(x|\beta^{-1})=E(x|\beta^{-1})=\beta e^{-\beta x}$, $\beta>0$, $x\ge0$. 

Sufficient statistics: $\text{E}[x]=\frac{1}{\beta}$, $\text{Var}[x]=\frac{1}{\beta^2}$.

Important property: Memorylessness: $P(X>x_1+x_2|X>x_1)=P(X>x_2)$.

Application: Model time between two events in a Poisson process.<br><br>

$\textbf{Gamma Distribution}$<br><br>
Probability density function: $p(x|\alpha,\beta)=\text{G}(x|\alpha,\beta)=\frac{\beta^{\alpha}}{\Gamma(\alpha)} x^{\alpha-1} e^{-\beta x}$, $\alpha>0$, $\beta>0$.

Sufficient statistics: $\text{E}[x]=\frac{\alpha}{\beta}$, $\text{Var}[x]=\frac{\alpha}{\beta^2}$.\\ 

Important property: If independent random variables $X_i \sim \text{G}(\alpha_i, \beta)$ for $i=1,\cdots,n$, then $\sum_{i=1}^n X_i \sim \text{G}(\sum_{i=1}^n \alpha_i, \beta)$.

Application: Often used as conjugate prior of Poisson and exponential distributions.<br><br>

$\textbf{Bonus problems}$<br><br>
Multivariate normal distribution: The $n$-dimensional random vector $X$ is normally distributed with mean vector
    $\mu = [\mu_1,\mu_2,\cdots,\mu_n]^T$ and covariance matrix $V, V_{ij} = cov(X_i, X_j)$, written $X\sim N(\mu,V)$, if
$f_X(x)=\frac{1}{\sqrt{(2\pi)^n|V|}}e^{-\frac{1}{2}(x-\mu)^T V^{-1} (x-\mu)}$. Where $|V|$ is the determinent of $V$.
The standard multivariate normal is defined for $\mu = [0,0,\cdots,0]^T, V = I_n$. Here, we are interested in writing a generic multivariate normal distribution in terms of the standard multivariate
normal.<br><br>
To this end, if $N\sim N(0,I_n)$ and $N_g\sim\ N(\mu,V)$, it can be shown that<br><br>
$N_g = \mu + EN$<br><br>
where $EE^T = V$. $N_1\sim N(\mu,V)$ where $\mu = [0.1,0.5]^T$ and $V = [[1.0,0.2],[0.2,1.0]]$. Using the equation above, write a python program in which $N_1$ is defined in terms of $N\sim N(0,I_n)$. Furthermore, for a comparison, plot the contour and the 3D representation of $N_1$ defined in both ways. Hint: to obtain the matrix $E$ from $V = EE^T$, you can use a Cholesky decomposition.


```python
mu=np.array([0.1, 0.5])
V=np.array([[1.0,0.2],[0.2,1.0]])
E=np.linalg.cholesky(V) # Cholesky decomposition
print("E=\n{}".format(E))

VR=np.dot(E,E.transpose())
print("VR=EE^T=\n{}".format(VR))

### First method: N1 representation: N1 ~ Normal(mu,V)
x1, x2 = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x1, x2))
N1 = multivariate_normal(mu, V) 
p=N1.pdf(pos) # probability density from N1
# contour
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.contourf(x1, x2, p)
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.title("Contour using the first method")
# 3D prepresentation
fig2 = plt.figure()
ax2 = Axes3D(fig2)
ax2.plot_surface(x1, x2, p)
plt.xlim(-3,3)
plt.ylim(-3,3)
ax2.set_zlim(0,0.18)
plt.title("3D prepresentation using the first method")
plt.show()


### Second method: N representation: N1 = mu + E N, N ~ normal(0,I)
x1, x2 = np.mgrid[-3:3:.01, -3:3:.01]
N = multivariate_normal([0, 0], [[1.0, 0], [0, 1.0]])
pos2=np.dstack((x1, x2)) 
p=N.pdf(pos2) # probability density from N, important: p_N1(N1) = p_N1(mu+EN) = p_N(N)
nr=pos2.shape[0]
nc=pos2.shape[1]
# convert coordinates from N to N1
pos3=np.zeros(shape=(nr,nc,2))
for r in range(nr):
    for c in range(nc):
        pos3[r,c,:]= mu + np.dot(E,pos2[r,c,:])
x1=pos3[:,:,0]
x2=pos3[:,:,1]
# contour
fig3 = plt.figure()
ax3 = fig3.add_subplot(111)
ax3.contourf(x1, x2,  p )
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.title("Contour using the second method")
# 3D prepresentation
fig4 = plt.figure()
ax4 = Axes3D(fig4)
ax4.plot_surface(x1, x2, p)
plt.xlim(-3,3)
plt.ylim(-3,3)
ax2.set_zlim(0,0.18)
plt.title("3D prepresentation using the second method")
plt.show()
```

Firstly, function np.linalg.cholesky() was used to decompose $V$ to $EE^T$. In first method: two variables $X_1$, $X_2$ were defined and varable $pos$ was canculated by using function np.dstack(($X_1$, $X_2$ )). Then random variable $N_1$ was obtained by using function multivariate_normal($\mu$,$V$), where $N_1 \sim\ N(\mu,V)$. Probability density $P$ from N1 was calculated and the contour and 3D prepresentation were plotted by using these there variables. In second method: variables $X_1$, $X_2$ and $N$ were defined, where $N \sim\ N(0,I)$. Probability density $P$ from N was obtained. Then $X_1$, $X_2$ were recalculated by using formula $N_g = \mu + EN$. Finally, the new contour and 3D prepresentation were plotted by using these new variables.
